```python
import matplotlib.pyplot as plt

def memory_histogram():
    # Data values provided
    data = [2,
    5,
    5,
    4,
    7,
    3,
    1,
    4,
    2,
    3,
    6,
    5,
    4,
    4,
    4,
    1,
    4,
    7,
    5,
    2,
    2,
    6,
    3,
    4,
    5,
    3,
    2,
    7,
    6,
    4,
    4,
    4,
    2,
    4,
    1,
    3,
    7,
    6,
    1,
    4,
    3
    ]
    
    # Creating a histogram
    plt.figure(figsize=(10, 6))
    plt.hist(data, bins=15, color='skyblue', edgecolor='black')
    plt.title('Histogram of Memory test results')
    plt.xlabel('Final Scores')
    plt.ylabel('Frequency')
    plt.show()
```

```python
def histogram_outliers_memory():
    import numpy as np
    # Re-running the entire process in one cell to avoid scope issues
    
    # Calculate the quartiles and interquartile range
    q1, q3 = np.percentile(data, [25, 75])
    iqr = q3 - q1
    
    # Determine the outlier bounds
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    
    # Find outliers
    outliers = [x for x in data if x < lower_bound or x > upper_bound]
    
    # Plotting the histogram with highlighted outliers
    plt.figure(figsize=(10, 6))
    
    # Plotting the regular data
    plt.hist([x for x in data if x not in outliers], bins=15, color='skyblue', edgecolor='black')
    
    # Plotting the outliers in a different color
    plt.hist(outliers, bins=15, color='red', edgecolor='black')
    
    plt.title('Histogram of Memory test results with Outliers Highlighted')
    plt.xlabel('Final Scores')
    plt.ylabel('Frequency')
    
    # Adding lines for lower and upper bounds
    plt.axvline(lower_bound, color='green', linestyle='dashed', linewidth=1)
    plt.axvline(upper_bound, color='green', linestyle='dashed', linewidth=1)
    
    plt.show()
```
